```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python

url1 = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code='
num = 85578   ### 얘를 바꿔주면 댐
url2 = '&type=after&onlyActualPointYn=Y&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page='
page = 1   ### 한 페이지당 평점 10개


response = req.urlopen(url1+str(num)+url2+str(page))
print(response)
```

```python
soup = BeautifulSoup(response, 'html.parser')   #추출
score_result = soup.find('div', class_='score_result')   # 리뷰와 평점이 있는 부분만 자르기
lis = score_result.find_all('li')    # li 별로 잘라서 벡터형태로 변경
```

```python
### 페이지 한 창의 평점 크롤링

for i in range(10):
    page=i
    li=lis[i]
    score = li.select_one("div.star_score > em").get_text()
    reple = li.select_one("div.score_reple > p").get_text()
    print('평점 :',  score)
    print('리뷰 :', reple, '\n')
```

```python
### 4가지 종류의 영화를 댓글창 10페이지까지 돌리면....


url1 = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code='
url2 = '&type=after&onlyActualPointYn=Y&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page='
movies = [85578, 136898]


for k in movies:
    num = k
    response = req.urlopen('https://movie.naver.com/movie/bi/mi/point.nhn?code='+str(num))
    soup = BeautifulSoup(response, 'html.parser')
    title= soup.select_one("title").string
    print('###', title, '###', '\n')
    for j in range(1, 5):
        page=j
        response = req.urlopen(url1+str(num)+url2+str(page))
        soup = BeautifulSoup(response, 'html.parser')   #추출
        score_result = soup.find('div', class_='score_result')   # 리뷰와 평점이 있는 부분만 자르기
        lis = score_result.find_all('li')
        for i in range(10):
            li=lis[i]
            score = li.select_one("div.star_score > em").string
            reple = li.select_one("div.score_reple > p").get_text()
            comment=[score, reple]
            print(comment, '\n')
```

```python
### 텍스트파일로 저ㅋ장ㅋ

url1 = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code='
url2 = '&type=after&onlyActualPointYn=Y&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page='
movies = [85578, 136898, 172454, 168298, 125494, 158885]

f=open('Comments.txt', 'w')
for k in movies:
    num = k
    response = req.urlopen('https://movie.naver.com/movie/bi/mi/point.nhn?code='+str(num))
    soup = BeautifulSoup(response, 'html.parser')
    title= soup.select_one("title").string
    f.write('\n'+'### '+title+' ###'+'\n')
    for j in range(1, 11):
        page=j
        response = req.urlopen(url1+str(num)+url2+str(page))
        soup = BeautifulSoup(response, 'html.parser')   #추출
        score_result = soup.find('div', class_='score_result')   # 리뷰와 평점이 있는 부분만 자르기
        lis = score_result.find_all('li')
        for i in range(10):
            li=lis[i]
            score = li.select_one("div.star_score > em").string
            reple = li.select_one("div.score_reple > p").get_text()
            comment=score+' , '+reple+'\n'

            f.write(comment)
f.close
```
