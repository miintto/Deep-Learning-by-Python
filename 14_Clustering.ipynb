# 14 Clustering

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

### Make blobs

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(random_state=1)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
_ = plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
### Kmeans 적용

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
print(kmeans.labels_)
print(kmeans.predict(X))
print(kmeans.cluster_centers_)
```

```python
colors = ['navy', 'turquoise', 'darkorange']
for color, i in zip(colors, [0, 1, 2]):
    plt.scatter(X[kmeans.labels_==i, 0], X[kmeans.labels_== i, 1], color=color)
_ = plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='^')
```

```python
### Agglomerative Clustering

average_linkage = AgglomerativeClustering(n_clusters=3, linkage='average', affinity='euclidean')
    ### affinity : 거리를 2-norm 으로 계산

average_linkage.fit(X)
```

```python
average_linkage.labels_
```

```python
for color, i in zip(colors, [0, 1, 2]):
    plt.scatter(X[average_linkage.labels_==i, 0], X[average_linkage.labels_== i, 1], c=color)
```
