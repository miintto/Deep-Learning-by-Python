### Train Set 수치화

```python
train = pd.read_csv("train.csv")   ### 불러오기
```

```python
train.head()
```

```python
print(train.dtypes)
print(np.shape(train))
```

```python
Pclass_cat = np_utils.to_categorical(train['Pclass'])
train['Pclass_1'] = Pclass_cat[:, 1].astype(int)
train['Pclass_2'] = Pclass_cat[:, 2].astype(int)
train['Pclass_3'] = Pclass_cat[:, 3].astype(int)
```

```python
### Sex 변환, male / female 

for i in range(np.shape(train)[0]):
    if(train.loc[i, 'Sex'] == 'male'):
        train.loc[i, 'Sex'] = 0
    elif(train.loc[i, 'Sex'] == 'female'):
        train.loc[i, 'Sex'] = 1
```

```python
### SibSp 변환 0 / 1~3 / 4~ 

for i in range(np.shape(train)[0]):
    if(train.loc[i, 'SibSp'] == 0):
        train.loc[i, 'SibSp'] = 0
    elif(train.loc[i, 'SibSp'] < 3):
        train.loc[i, 'SibSp'] = 1
    else:
        train.loc[i, 'SibSp'] = 2

SibSp_cat = np_utils.to_categorical(train['SibSp'])
train['SibSp_0'] = SibSp_cat[:, 0].astype(int)
train['SibSp_1'] = SibSp_cat[:, 1].astype(int)
train['SibSp_2'] = SibSp_cat[:, 2].astype(int)
```

```python
### Parch 변환 0 / 1~3 / 4~ 

for i in range(np.shape(train)[0]):
    if(train.loc[i, 'Parch'] == 0):
        train.loc[i, 'Parch'] = 0
    elif(train.loc[i, 'Parch'] < 4):
        train.loc[i, 'Parch'] = 1
    else:
        train.loc[i, 'Parch'] = 2

Parch_cat = np_utils.to_categorical(train['Parch'])
train['Parch_0'] = Parch_cat[:, 0].astype(int)
train['Parch_1'] = Parch_cat[:, 1].astype(int)
train['Parch_2'] = Parch_cat[:, 2].astype(int)
```

```python
### Fare 변환

train['Fare'] = np.sqrt(train['Fare'])
train.loc[train['Fare']>9, 'Fare'] = 9
train['Fare'] = (train['Fare']-np.mean(train['Fare']))/np.std(train['Fare'])
```

```python
### Embarked 변환, C / Q / S

for i in range(np.shape(train)[0]):
    if(train.loc[i, 'Embarked'] == 'C'):
        train.loc[i, 'Embarked'] = 0
    elif(train.loc[i, 'Embarked'] == 'Q'):
        train.loc[i, 'Embarked'] = 1
    elif(train.loc[i, 'Embarked'] == 'S'):
        train.loc[i, 'Embarked'] = 2
    else:
        train.loc[i, 'Embarked'] = 3
        
Embarked_cat = np_utils.to_categorical(train['Embarked'])
train['Embarked_C'] = Embarked_cat[:, 0].astype(int)
train['Embarked_Q'] = Embarked_cat[:, 1].astype(int)
train['Embarked_S'] = Embarked_cat[:, 2].astype(int)
```

```python
### emabark 결측치는 둘다 S로 분류 

train.loc[train.loc[:,'Embarked_C']+train.loc[:,'Embarked_Q']+train.loc[:,'Embarked_S']==0, 'Embarked_S'] = 1
```

```python
### Survived, Pclass, Sex, SibSp, Embarked 만 남기고 모두 삭제

train = train.drop('PassengerId', axis=1)
train = train.drop('Pclass', axis=1)
train = train.drop('Name', axis=1)
train = train.drop('SibSp', axis=1)
train = train.drop('Parch', axis=1)
train = train.drop('Ticket', axis=1)
train = train.drop('Cabin', axis=1)
train = train.drop('Embarked', axis=1)
```

### Model 설계

```python
import numpy as np
import pandas as pd
```

```python
dat = pd.read_csv("train_fix.csv")   ### 데이터 불러오기
```

```python
dat = dat.drop('Embarked_C', axis=1)
dat = dat.drop('Embarked_Q', axis=1)
dat = dat.drop('Embarked_S', axis=1)
dat = dat.drop('Age', axis=1)
dat.head(10)
```

```python
Y_train = dat['Survived']
X_train = dat.drop('Survived', axis=1)
```

```python
X_train = np.array(X_train)
Y_train = np.array(Y_train).reshape(-1, 1)

print(X_train[:5])
print(Y_train[:5])
print(np.shape(X_train))
print(np.shape(Y_train))
```

```python
import tensorflow as tf
sess = tf.Session()
from sklearn.utils import shuffle
```

```python
n_lay1 = 20   ### hidden layer 1
n_lay2 = 5   ### hidden layer 2
n_lay3 = 2   ### hidden layer 2


tf.set_random_seed(1234)

x = tf.placeholder(tf.float32, shape=[None, np.shape(X_train)[1]])
t = tf.placeholder(tf.float32, shape=[None, 1])
keep_prob = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.truncated_normal([np.shape(X_train)[1], n_lay1], seed = 0))
b1 = tf.Variable(tf.zeros([n_lay1]))
z1 = tf.nn.dropout(tf.nn.relu(tf.matmul(x, w1)+b1), keep_prob)

w2 = tf.Variable(tf.truncated_normal([n_lay1, n_lay2], seed = 10))
b2 = tf.Variable(tf.zeros([n_lay2]))
z2 = tf.nn.dropout(tf.nn.relu(tf.matmul(z1, w2)+b2), keep_prob)

w3 = tf.Variable(tf.truncated_normal([n_lay2, 1], seed = 20))
b3 = tf.Variable(tf.zeros([1]))
y = tf.nn.sigmoid(tf.matmul(z2, w3)+b3)   ### Output

cross_ent = -tf.reduce_mean(t*tf.log(y)+(1-t)*tf.log(1-y))   ### Error function 계산

lr = 0.1   ### learning rate = 0.03
train_step = tf.train.GradientDescentOptimizer(lr).minimize(cross_ent)

correct_prediction = tf.equal(tf.to_float(tf.greater(y, 0.5)), t)   ### y와 0.5의 값 비교 ( T, F 반환 )
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))   ### 정확도 측정

init = tf.global_variables_initializer()   ### initialize 필요
sess.run(init)
```

### Train

```python
n_samples = X_train.shape[0]
bat_size = 16   ### 보통 2^n 꼴로 정한다.
n_batch = n_samples // bat_size

print(n_samples)   ### 
print(n_batch)   ### 
```

```python
for epoch in range(100):   ### 100번 계산한다.
    X_, Y_ = shuffle(X_train, Y_train)   ### 순서를 섞어준다
    
    for i in range(n_batch):
        sess.run(train_step, feed_dict={x:X_[i*bat_size:i*bat_size+bat_size], t:Y_[i*bat_size:i*bat_size+bat_size], keep_prob:0.7})
    
    if epoch % 10 == 0:   ### 이건 그냥 잘 돌아가는지 판단하려고...
        loss_train = cross_ent.eval(session=sess, feed_dict={x:X_train, t:Y_train, keep_prob:1.0})
        accuracy_train = accuracy.eval(session=sess, feed_dict={x:X_train, t:Y_train, keep_prob:1.0})
        print("epoch =", epoch, ", Loss =", loss_train, ", Accuracy =", accuracy_train)
    if pd.isnull(loss_train):
        break
```

```python
loss_train = cross_ent.eval(session=sess, feed_dict={x:X_train, t:Y_train, keep_prob:1})
accuracy_train = accuracy.eval(session=sess, feed_dict={x:X_train, t:Y_train, keep_prob:1})
print(loss_train, accuracy_train)   ### loss < 0.346 까지...
```
