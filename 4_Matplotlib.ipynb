# Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
```

### Graph (line chart)

```python
years=np.arange(1950, 2020, 10)
gdp=np.array([300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3])
print(years)
print(gdp)
```

```python
plt.plot(years, gdp, color='r', marker='.', linestyle='-')
    # marker = 점 모양 ('.' ',' 'o' 'v' 'x')
    # linestyle = 선 모양 ('-' '--' '-.' ':')
plt.title('Normal GDP')
plt.ylabel('Billoin $')
plt.xlabel('years')
```

```python
#### color, marker, linstlye 을 차례로 입력해도 똑같은 결과가 나옴.
#### 즉, color = 'g', marker = 'o', linstlye = '-' 를 'go-' 로 해도 나옴.

plt.plot(years, gdp, 'r.-')   # 똑같이 나옴
plt.title('Normal GDP')
plt.ylabel('Billoin $')
plt.xlabel('years')
```

```python
variance=[1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared=[256, 128, 64, 32, 16, 8, 4 , 2, 1]
total_error=[x+y for x, y in zip(variance, bias_squared)]
```

```python
xs=[i for i, _ in enumerate(variance)]
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-', label='bias^2')
plt.plot(xs, total_error, 'b:', label='tatal error')

plt.legend(loc=9)
plt.title("The Bias-Variance Trade-off")
_ =plt.xlabel('model complexity')
```

### Bar Chart

### Histogram

```python
from collections import Counter

decile=lambda grade: grade // 10*10
histogram=Counter(decile(grade) for grade in grades)

print(histogram)
print(histogram.keys())
```

```python
plt.bar([x for x in histogram.keys()], histogram.values(), 4)
plt.axis([-5, 105, 0, 5])

plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('Number of Students')
plt.title('Distribution of Exam 1 Grades')
```

### Scatterplots

```python
friends=[70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes=[175, 170, 205, 120, 220, 130, 105, 145, 190]
labels=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends, minutes)
for label, fnd_cnt, min_cnt in zip(labels, friends, minutes):
    plt.annotate(label, 
                xy=(fnd_cnt, min_cnt),    # 해당 좌표
                xytext=(5, -5),    # 문자 넣을 좌표 (해당 좌표로부터 거리)
                textcoords='offset points')    # 설정값인듯...
plt.title('Daily Minute vs. Number of friends')
plt.xlabel('number of friends')
plt.ylabel('daily minute spent on the site')
```
